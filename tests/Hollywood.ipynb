```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact

# Exemple de données fictives
# Population générale
population_data = {
    "Ethnicity": ["African American", "American Indian", "Arab", "Asian", "Latino", "Jewish", "Caucasian"],
    "Proportion": [13.4, 1.3, 0.6, 5.9, 18.5, 2.0, 58.3]
}

# Acteurs de films (fictifs, ajustez selon vos données)
actors_data = {
    "Ethnicity": ["African American", "American Indian", "Arab", "Asian", "Latino", "Jewish", "Caucasian"],
    "Proportion": [10.0, 0.5, 1.2, 6.8, 12.3, 1.5, 67.7]
}

# Convertir les données en DataFrames
population_df = pd.DataFrame(population_data)
actors_df = pd.DataFrame(actors_data)

# Fonction pour afficher le graphique interactif
def plot_ethnicity_comparison(genre="All", start_year=1950, end_year=2022):
    # Filtrer les données des acteurs selon le genre et la période (à ajuster avec vos colonnes réelles)
    filtered_actors_df = actors_df  # Remplacez par vos données filtrées selon genre et période
    
    # Préparer les données pour le graphique
    ethnicities = population_df["Ethnicity"]
    population_proportions = population_df["Proportion"]
    actor_proportions = filtered_actors_df["Proportion"]
    
    # Configuration du graphique
    fig, ax = plt.subplots(figsize=(10, 6))
    
    bar_width = 0.4
    x = range(len(ethnicities))
    
    # Histogramme de gauche pour la population générale
    ax.barh(
        x, 
        population_proportions, 
        height=bar_width, 
        color='skyblue', 
        label='Population générale', 
        align='center'
    )
    
    # Histogramme de droite pour les acteurs
    ax.barh(
        x, 
        [-p for p in actor_proportions],  # Négatif pour afficher à droite
        height=bar_width, 
        color='salmon', 
        label='Acteurs de films', 
        align='center'
    )
    
    # Ajouter des annotations
    for i, (pop, act) in enumerate(zip(population_proportions, actor_proportions)):
        ax.text(pop + 1, i, f'{pop}%', va='center', ha='left', color='blue')
        ax.text(-act - 1, i, f'{act}%', va='center', ha='right', color='red')
    
    # Configurer les axes
    ax.set_yticks(x)
    ax.set_yticklabels(ethnicities)
    ax.axvline(0, color='black', linewidth=0.8)  # Ligne centrale
    ax.set_xlabel("Proportion (%)")
    ax.set_title(f"Comparaison des proportions d'ethnicités (Genre: {genre}, Période: {start_year}-{end_year})")
    ax.legend(loc='upper center')
    
    plt.tight_layout()
    plt.show()

# Créer un widget interactif
interact(
    plot_ethnicity_comparison,
    genre=["All", "Action", "Drama", "Comedy", "Thriller"],  # Liste des genres possibles
    start_year=(1950, 2022, 1),  # Slider pour l'année de début
    end_year=(1950, 2022, 1)     # Slider pour l'année de fin
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact

# Function to calculate proportions of actor ethnicities
def calculate_actor_ethnicities(df, period, genre):
    # Filter by period
    if period != "All":
        start_year, end_year = map(int, period.split('-'))
        df = df[df['release_y'].between(start_year, end_year)]
    
    # Filter by genre
    if genre != "All":
        df = df[df['main_genre'] == genre]
    
    # Calculate proportions
    ethnicity_counts = df['actor_ethnicity_classification'].value_counts(normalize=True)
    return ethnicity_counts.reindex(ethnicities, fill_value=0)

# Function to plot the comparison graph
def plot_ethnicity_comparison(period="All", genre="All"):
    # Get population proportions for the selected period
    if period == "All":
        population_ethnicities = hollywood_ethnic_realworld.groupby('group')['size'].mean()
    else:
        population_ethnicities = hollywood_ethnic_realworld[
            hollywood_ethnic_realworld['new_period'] == period
        ].set_index('group')['size']
    
    # Get actor proportions for the selected period and genre
    actor_ethnicities = calculate_actor_ethnicities(hollywood_data_ethnicity, period, genre)

    # Prepare data for the graph
    left_values = population_ethnicities.values * 100  # Convert to percentages
    right_values = actor_ethnicities.values * 100
    ethnicities_labels = population_ethnicities.index

    # Create the graph
    fig, ax = plt.subplots(figsize=(12, 8))
    
    bar_width = 0.4
    y_positions = range(len(ethnicities_labels))

    # Left bar for population
    ax.barh(
        y_positions, 
        left_values, 
        height=bar_width, 
        color='skyblue', 
        label="Population générale", 
        align='center'
    )

    # Right bar for actors
    ax.barh(
        y_positions, 
        -right_values,  # Negative for right side
        height=bar_width, 
        color='orange', 
        label="Acteurs de films", 
        align='center'
    )

    # Add annotations
    for i, (left, right) in enumerate(zip(left_values, right_values)):
        ax.text(left + 0.5, i, f"{left:.1f}%", va='center', ha='left', color='blue')
        ax.text(-right - 0.5, i, f"{right:.1f}%", va='center', ha='right', color='red')

    # Configure graph
    ax.axvline(0, color='black', linewidth=0.8)  # Central line
    ax.set_yticks(y_positions)
    ax.set_yticklabels(ethnicities_labels)
    ax.set_xlabel("Proportion (%)")
    ax.set_title(f"Comparaison des proportions d'ethnicités (Période: {period}, Genre: {genre})")
    ax.legend(loc='upper center')

    plt.tight_layout()
    plt.show()

# Ethnicities list
ethnicities = [
    "African Americans", 
    "American Indians", 
    "Arab Americans", 
    "Asian Americans", 
    "Caucasian Americans", 
    "Jewish Americans", 
    "Latino Americans"
]

# Interactive widget for the graph
interact(
    plot_ethnicity_comparison,
    period=["All", "1950-1965", "1966-1980", "1981-1995", "1996-2012"],  # Time periods
    genre=["All", "Action/Adventure", "Drama", "Comedy", "Thriller/Suspense", 
           "Horror", "Science Fiction (Sci-Fi)", "Fantasy", "Romance", 
           "Documentary", "Animation/Family"]  # Genres
)
```

```python
# avec la moyenne ponderee 

import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact

# Fonction pour calculer les proportions d'ethnicités par film
def calculate_weighted_actor_ethnicities(df, period, genre):
    # Filtrer par période
    if period != "All":
        start_year, end_year = map(int, period.split('-'))
        df = df[df['release_y'].between(start_year, end_year)]
    
    # Filtrer par genre
    if genre != "All":
        df = df[df['main_genre'] == genre]
    
    # Calculer la taille du casting pour chaque film
    df['casting_size'] = df.groupby('wiki_movie_id')['wiki_movie_id'].transform('count')
    
    # Calculer les proportions d'ethnicités pour chaque film
    film_ethnicities = df.groupby(['wiki_movie_id', 'actor_ethnicity_classification']).size().unstack(fill_value=0)
    film_ethnicities = film_ethnicities.div(film_ethnicities.sum(axis=1), axis=0)  # Proportions par film
    
    # Ajouter la taille du casting
    film_ethnicities['casting_size'] = df.groupby('wiki_movie_id')['casting_size'].first()

    # Moyenne pondérée des proportions
    weighted_ethnicities = film_ethnicities.drop(columns='casting_size').multiply(film_ethnicities['casting_size'], axis=0).sum() / film_ethnicities['casting_size'].sum()
    
    # Remplir les valeurs manquantes avec 0 pour les ethnicités non représentées
    weighted_ethnicities = weighted_ethnicities.reindex(ethnicities, fill_value=0)
    
    return weighted_ethnicities

# Fonction pour afficher le graphe
def plot_ethnicity_comparison(period="All", genre="All"):
    # Obtenir les proportions de la population réelle pour la période sélectionnée
    if period == "All":
        population_ethnicities = hollywood_ethnic_realworld.groupby('group')['size'].mean()
    else:
        population_ethnicities = hollywood_ethnic_realworld[
            hollywood_ethnic_realworld['new_period'] == period
        ].set_index('group')['size']
    
    # Obtenir les proportions des acteurs pour la période et le genre sélectionnés
    actor_ethnicities = calculate_weighted_actor_ethnicities(hollywood_data_ethnicity, period, genre)

    # Préparer les données pour le graphe
    left_values = population_ethnicities.values * 100  # Convertir en pourcentages
    right_values = actor_ethnicities.values * 100
    ethnicities_labels = population_ethnicities.index

    # Créer le graphique
    fig, ax = plt.subplots(figsize=(12, 8))
    
    bar_width = 0.4
    y_positions = range(len(ethnicities_labels))

    # Barres pour la population générale (gauche)
    ax.barh(
        y_positions, 
        left_values, 
        height=bar_width, 
        color='skyblue', 
        label="Population générale", 
        align='center'
    )

    # Barres pour les acteurs (droite)
    ax.barh(
        y_positions, 
        -right_values,  # Négatif pour les afficher à droite
        height=bar_width, 
        color='orange', 
        label="Acteurs de films (moyenne pondérée)", 
        align='center'
    )

    # Ajouter des annotations
    for i, (left, right) in enumerate(zip(left_values, right_values)):
        ax.text(left + 0.5, i, f"{left:.1f}%", va='center', ha='left', color='blue')
        ax.text(-right - 0.5, i, f"{right:.1f}%", va='center', ha='right', color='red')

    # Configurer le graphique
    ax.axvline(0, color='black', linewidth=0.8)  # Ligne centrale
    ax.set_yticks(y_positions)
    ax.set_yticklabels(ethnicities_labels)
    ax.set_xlabel("Proportion (%)")
    ax.set_title(f"Comparaison des proportions d'ethnicités (Période: {period}, Genre: {genre})")
    ax.legend(loc='upper center')

    plt.tight_layout()
    plt.show()

# Liste des ethnicités
ethnicities = [
    "African Americans", 
    "American Indians", 
    "Arab Americans", 
    "Asian Americans", 
    "Caucasian Americans", 
    "Jewish Americans", 
    "Latino Americans"
]

# Interactif
interact(
    plot_ethnicity_comparison,
    period=["All", "1950-1965", "1966-1980", "1981-1995", "1996-2012"],  # Périodes temporelles
    genre=["All", "Action/Adventure", "Drama", "Comedy", "Thriller/Suspense", 
           "Horror", "Science Fiction (Sci-Fi)", "Fantasy", "Romance", 
           "Documentary", "Animation/Family"]  # Genres
)
```

```python
# Ensure Time Period is the index for real-world proportions
male_hollywood_realworld_proportions.set_index('Time Period', inplace=True)
female_hollywood_realworld_proportions.set_index('Time Period', inplace=True)

# Ensure release_y is an integer for hollywood_data
hollywood_data['release_y'] = hollywood_data['release_y'].astype(int)

# Function to calculate gender proportions
def calculate_gender_proportions(df, period, genres):
    # Filter by period
    if period != "All":
        start_year, end_year = map(int, period.split('-'))
        df = df[df['release_y'].between(start_year, end_year)]
    
    # Filter by genres
    if genres != "All":
        df = df[df['main_genre'].isin(genres)]
    
    # Make a copy to avoid SettingWithCopyWarning
    df = df.copy()

    # Calculate casting size
    df['casting_size'] = df.groupby('wiki_movie_id')['wiki_movie_id'].transform('count')
    
    # Calculate gender proportions per film
    gender_counts = df.groupby(['wiki_movie_id', 'actor_gender']).size().unstack(fill_value=0)
    gender_proportions = gender_counts.div(gender_counts.sum(axis=1), axis=0)
    
    # Add casting size
    gender_proportions['casting_size'] = df.groupby('wiki_movie_id')['casting_size'].first()

    # Weighted mean proportions
    weighted_proportions = gender_proportions.drop(columns='casting_size').multiply(
        gender_proportions['casting_size'], axis=0).sum() / gender_proportions['casting_size'].sum()
    
    # Return male and female proportions
    male_proportion = weighted_proportions.get('M', 0)  # Default to 0 if not found
    female_proportion = weighted_proportions.get('F', 0)  # Default to 0 if not found
    return male_proportion, female_proportion

# Function to plot gender proportions
def plot_gender_proportions(period="All", genres=["All"]):
    # List of genres
    genre_categories = genres if genres != ["All"] else hollywood_data['main_genre'].unique()

    # Prepare data for each genre
    proportions = []
    for genre in genre_categories:
        male, female = calculate_gender_proportions(hollywood_data, period, [genre])
        proportions.append({"category": genre, "male": male, "female": female})

    # Add real-world population proportions
    if period == "All":
        male_population = male_hollywood_realworld_proportions.mean(axis=1).mean()
        female_population = female_hollywood_realworld_proportions.mean(axis=1).mean()
    else:
        male_population = male_hollywood_realworld_proportions.loc[period].mean()
        female_population = female_hollywood_realworld_proportions.loc[period].mean()
    
    proportions.append({"category": "Population réelle", "male": male_population, "female": female_population})

    # Create DataFrame for the graph
    proportions_df = pd.DataFrame(proportions)

    # Create the graph
    fig, ax = plt.subplots(figsize=(12, 6))

    # Stacked bars
    ax.bar(
        proportions_df['category'], 
        proportions_df['male'], 
        color='skyblue', 
        label='Hommes'
    )
    ax.bar(
        proportions_df['category'], 
        proportions_df['female'], 
        bottom=proportions_df['male'], 
        color='salmon', 
        label='Femmes'
    )

    # Configure the graph
    ax.set_title(f"Proportions Hommes/Femmes par Catégorie (Période: {period})")
    ax.set_ylabel("Proportion")
    ax.set_xticks(range(len(proportions_df['category'])))
    ax.set_xticklabels(proportions_df['category'], rotation=45, ha='right')
    ax.legend()

    plt.tight_layout()
    plt.show()
```

```python
# Fonction pour calculer les proportions homme/femme par genre et population réelle
def calculate_gender_proportions(df, period, genres):
    # Filtrer par période
    if period != "All":
        start_year, end_year = map(int, period.split('-'))
        df = df[df['release_y'].between(start_year, end_year)]
    
    # Filtrer par genres
    if genres != "All":
        df = df[df['main_genre'].isin(genres)]
    
    # Calculer la taille du casting par film
    df = df.copy()  # Éviter le SettingWithCopyWarning
    df['casting_size'] = df.groupby('wiki_movie_id')['wiki_movie_id'].transform('count')
    
    # Calculer les proportions homme/femme par film
    gender_counts = df.groupby(['wiki_movie_id', 'actor_gender']).size().unstack(fill_value=0)
    gender_proportions = gender_counts.div(gender_counts.sum(axis=1), axis=0)
    
    # Ajouter la taille du casting
    gender_proportions['casting_size'] = df.groupby('wiki_movie_id')['casting_size'].first()

    # Moyennes pondérées des proportions
    weighted_proportions = gender_proportions.drop(columns='casting_size').multiply(
        gender_proportions['casting_size'], axis=0).sum() / gender_proportions['casting_size'].sum()
    
    # Retourner les proportions homme et femme
    male_proportion = weighted_proportions.get('M', 0)  # Retourne 0 si "M" n'existe pas
    female_proportion = weighted_proportions.get('F', 0)  # Retourne 0 si "F" n'existe pas
    return male_proportion, female_proportion

# Fonction pour afficher le graphe
def plot_gender_proportions(period="All", genres=["All"]):
    # Liste des genres
    genre_categories = genres if genres != ["All"] else hollywood_data['main_genre'].unique()

    # Préparer les données pour chaque genre
    proportions = []
    for genre in genre_categories:
        male, female = calculate_gender_proportions(hollywood_data, period, [genre])
        proportions.append({"category": genre, "male": male, "female": female})

    # Ajouter les proportions de la population réelle
    if period == "All":
        male_population = male_hollywood_realworld_proportions.mean(axis=1).mean()
        female_population = female_hollywood_realworld_proportions.mean(axis=1).mean()
    else:
        male_population = male_hollywood_realworld_proportions.loc[period].mean()
        female_population = female_hollywood_realworld_proportions.loc[period].mean()
    
    proportions.append({"category": "Population réelle", "male": male_population, "female": female_population})

    # Créer le DataFrame pour le graphe
    proportions_df = pd.DataFrame(proportions)

    # Créer le graphique
    fig, ax = plt.subplots(figsize=(12, 6))

    # Barres empilées
    ax.bar(
        proportions_df['category'], 
        proportions_df['male'], 
        color='skyblue', 
        label='Hommes'
    )
    ax.bar(
        proportions_df['category'], 
        proportions_df['female'], 
        bottom=proportions_df['male'], 
        color='orange', 
        label='Femmes'
    )

    # Configurer le graphique
    ax.set_title(f"Proportions Hommes/Femmes par Catégorie (Période: {period})")
    ax.set_ylabel("Proportion")
    ax.set_xticks(range(len(proportions_df['category'])))
    ax.set_xticklabels(proportions_df['category'], rotation=45, ha='right')
    ax.legend()

    plt.tight_layout()
    plt.show()



# Genres disponibles
genres_list = [
    "All", "Action/Adventure", "Drama", "Comedy", "Thriller/Suspense", 
    "Horror", "Science Fiction (Sci-Fi)", "Fantasy", "Romance", 
    "Documentary", "Animation/Family"
]

# Interactif
interact(
    plot_gender_proportions,
    period=["All", "1950-1965", "1966-1980", "1981-1995", "1996-2012"],  # Périodes temporelles
    genres=[genres_list]  # Genres
)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

# Fonction pour calculer la distribution des âges des acteurs
def calculate_actor_age_distribution(df, period, genres, gender):
    # Filtrer par période
    if period != "All":
        start_year, end_year = map(int, period.split('-'))
        df = df[df['release_x'].between(start_year, end_year)]
    
    # Filtrer par genres
    if genres != "All":
        df = df[df['main_genre'] == genres]
    
    # Filtrer par gender
    if gender != "Both":
        df = df[df['actor_gender'] == gender]
    
    # Calculer les distributions d'âges
    age_distribution, bins = np.histogram(df['age_at_release'], bins=range(0, 105, 5), density=True)
    return age_distribution, bins

# Fonction pour récupérer la distribution de la population réelle
def get_realworld_age_distribution(period, gender):
    if gender == "Male":
        realworld_data = male_hollywood_realworld_averages
    elif gender == "Female":
        realworld_data = female_hollywood_realworld_averages
    else:  # Both
        realworld_data = bothsexes_hollywood_realworld_averages

    # Extraire la ligne correspondant à la période
    if period == "All":
        # Moyenne uniquement sur les colonnes numériques
        distribution = realworld_data.select_dtypes(include=[np.number]).mean(axis=0).values
    else:
        # Extraire la ligne correspondant à la période
        distribution = realworld_data.loc[period].select_dtypes(include=[np.number]).values

    # Normaliser la distribution
    normalized_distribution = distribution / distribution.sum()
    return normalized_distribution, realworld_data.columns[1:].astype(str)

# Fonction pour tracer les graphes comparatifs
def plot_age_distribution(period="All", genres="All", gender="Both"):
    # Obtenir la distribution des acteurs
    actor_distribution, actor_bins = calculate_actor_age_distribution(hollywood_data, period, genres, gender)

    # Obtenir la distribution réelle
    realworld_distribution, age_bins = get_realworld_age_distribution(period, gender)

    # Tracer les deux distributions
    fig, ax = plt.subplots(figsize=(10, 6))
    width = 2  # Largeur des barres

    # Distribution des acteurs
    ax.bar(
        actor_bins[:-1] + width / 2, 
        actor_distribution, 
        width=width, 
        color='blue', 
        label="Acteurs (Films)"
    )

    # Distribution réelle
    ax.bar(
        np.arange(0, 105, 5) + width / 2, 
        realworld_distribution, 
        width=width, 
        color='orange', 
        alpha=0.7, 
        label="Population réelle"
    )

    # Configurer le graphique
    ax.set_title(f"Comparaison des distributions d'âges ({gender}, Période: {period}, Genre: {genres})")
    ax.set_xlabel("Âge")
    ax.set_ylabel("Proportion normalisée")
    ax.set_xticks(np.arange(0, 105, 5))
    ax.set_xticklabels([f"{i}-{i+4}" for i in range(0, 100, 5)] + ["100+"])
    ax.legend()
    plt.tight_layout()
    plt.show()

# Interactif
interact(
    plot_age_distribution,
    period=["All", "1950-1965", "1966-1980", "1981-1995", "1996-2012"],  # Périodes temporelles
    genres=["All", "Action/Adventure", "Drama", "Comedy", "Thriller/Suspense", 
            "Horror", "Science Fiction (Sci-Fi)", "Fantasy", "Romance", 
            "Documentary", "Animation/Family"],  # Genres
    gender=["Both", "Male", "Female"]  # Genre des acteurs
)
```
